```python
features = data.iloc[:, :4]
labels = data[['Species']]
le = LabelEncoder()
labels_encoded = le.fit_transform( labels )
#display( labels_encoded )
```

```python
X_train, X_test, y_train, y_test = train_test_split( features, labels_encoded, test_size = 0.33 )
display( X_train )
display( X_test )
display( y_train )
display( y_test )
```

```python
def model_tester( model, X_train, X_test, y_train, y_test ):
    #fitting the model
    model.fit( X_train, y_train )

    #predicting values
    y_predicted = model.predict( X_test )

    #measuring the accuracy
    accuracy = accuracy_score( y_test, y_predicted )

    #plotting the confusion matrix
    cm = confusion_matrix( y_test, y_predicted )
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, cmap="Reds", annot=cm, fmt="d", cbar=True, 
                linewidths=0.5)
    print( "Accuracy score of KNeighborsClassifier with 3 neighbors: ", accuracy_score( y_test, y_predicted) )
    plt.title(f"Confusion Matrix of {model} model")
    plt.xlabel("Predicted Label")
    plt.ylabel("True Label")
    plt.show()

    scores = precision_recall_fscore_support( y_test, y_predicted )
    print(scores[3])
```

```python
knn = KNeighborsClassifier( n_neighbors = 3 )
model_tester( knn, X_train, X_test, y_train, y_test )
```

```python
logistic_regression = LogisticRegression( multi_class = "multinomial", random_state = 42 )
model_tester( logistic_regression, X_train, X_test, y_train, y_test )
```
