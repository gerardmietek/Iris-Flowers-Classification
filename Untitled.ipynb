```python
import pandas as p
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display
```

```python
data = pd.read_csv( "Iris.csv" )
display( data.head() )
```

```python
print( "Number of rows: ", data.shape[0] )
print( "Number of columns: ", data.shape[1] )
```

```python
print( data.info() )
print( data["Species"].unique() )
```

```python
num_of_species = len( data["Species"].unique() )
print( f"There are {num_of_species} unique species in this dataset" )
```

```python
num_of_duplicates = data.duplicated().sum()
print( f"There is {num_of_duplicates} duplicated records in this dataset" )
```

```python
num_of_nulls = data.isnull().sum()
print( num_of_nulls )
print( "None of the features contains any null values" )
```

```python
print( data.groupby("Species").size() )
print( "\nEvery species has equal number of records that represent it" )
```

```python
display( data.iloc[:, 1:].describe().round(2) ) #excluded the id column that does not matter in statistical description
```

```python
data = data.iloc[:, 1:]
display( data.head() )
```

```python
fig, axes = plt.subplots( 2, 2, figsize=(10, 10) )

axes[0, 0].boxplot(data["SepalLengthCm"], 
                   patch_artist=True, 
                   boxprops=dict(facecolor='skyblue', color='blue'),
                   medianprops=dict(color='red', linewidth=2))
axes[0, 0].set_title("Sepal Length Boxplot", fontsize=14, fontweight='bold')
axes[0, 0].set_ylabel("Length (cm)", fontsize=12)
axes[0, 0].grid(True, linestyle='--', alpha=0.5)

axes[0, 1].boxplot(data["SepalWidthCm"], 
                   patch_artist=True, 
                   boxprops=dict(facecolor='lightgreen', color='green'),
                   medianprops=dict(color='red', linewidth=2))
axes[0, 1].set_title("Sepal Width Boxplot", fontsize=14, fontweight='bold')
axes[0, 1].set_ylabel("Width (cm)", fontsize=12)
axes[0, 1].grid(True, linestyle='--', alpha=0.5)

axes[1, 0].boxplot(data["PetalLengthCm"], 
                   patch_artist=True, 
                   boxprops=dict(facecolor='lightcoral', color='darkred'),
                   medianprops=dict(color='red', linewidth=2))
axes[1, 0].set_title("Petal Length Boxplot", fontsize=14, fontweight='bold')
axes[1, 0].set_ylabel("Length (cm)", fontsize=12)
axes[1, 0].grid(True, linestyle='--', alpha=0.5)

axes[1, 1].boxplot(data["PetalWidthCm"], 
                   patch_artist=True, 
                   boxprops=dict(facecolor='lightyellow', color='orange'),
                   medianprops=dict(color='red', linewidth=2))
axes[1, 1].set_title("Petal Width Boxplot", fontsize=14, fontweight='bold')
axes[1, 1].set_ylabel("Width (cm)", fontsize=12)
axes[1, 1].grid(True, linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots( 2, 2, figsize=(10, 10) )

axes[0, 0].hist(data["SepalLengthCm"], edgecolor = "blue", color = "skyblue"  )
axes[0, 0].set_title("Sepal Length distribution", fontsize=14, fontweight='bold')
axes[0, 0].set_ylabel("Count of flowers", fontsize=12)
axes[0, 0].set_xlabel("Length (cm)", fontsize=12)

axes[0, 1].hist(data["SepalWidthCm"], edgecolor = "green", color = "lightgreen"  )
axes[0, 1].set_title("Sepal Width distribution", fontsize=14, fontweight='bold')
axes[0, 1].set_ylabel("Count of flowers", fontsize=12)
axes[0, 1].set_xlabel("Width (cm)", fontsize=12)

axes[1, 0].hist(data["PetalLengthCm"], edgecolor = "darkred", color = "lightcoral"  )
axes[1, 0].set_title("Petal Length distribution", fontsize=14, fontweight='bold')
axes[1, 0].set_ylabel("Count of flowers", fontsize=12)
axes[1, 0].set_xlabel("Length (cm)", fontsize=12)

axes[1, 1].hist(data["PetalWidthCm"], edgecolor = "orange", color = "lightyellow"  )
axes[1, 1].set_title("Petal Width distribution", fontsize=14, fontweight='bold')
axes[1, 1].set_ylabel("Count of flowers", fontsize=12)
axes[1, 1].set_xlabel("Length (cm)", fontsize=12)

plt.tight_layout()
plt.show()
```

```python
correlation_matrix = data.iloc[:, :-1].corr()
plt.figure( figsize=(10, 10) )
sns.heatmap( correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5 )
plt.title( 'Correlation Heatmap of Iris Dataset Features', fontsize = 20, pad = 20 )
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```

```python
colors = {"Iris-setosa": "purple", "Iris-versicolor": "green", "Iris-virginica": "orange"}

fig, axes = plt.subplots(2, 2, figsize=(10, 10))

plt.subplot(2,2,1)
b1 = sns.boxplot( data = data, x = 'Species', y = 'SepalLengthCm', palette = colors, hue = 'Species', saturation = 15 )
b1.set_title( "Sepal Length for every species", fontsize=14, fontweight='bold' )
b1.set_xlabel( '' )
b1.set_ylabel( 'Length (cm)' )

plt.subplot(2,2,2)
b2 = sns.boxplot( data = data, x = 'Species', y = 'SepalWidthCm', palette = colors, hue = 'Species' )
b2.set_title( "Sepal Width for every species", fontsize=14, fontweight='bold' )
b2.set_ylabel( 'Width (cm)' )

plt.subplot(2,2,3)
b3 = sns.boxplot( data = data, x = 'Species', y = 'PetalLengthCm', palette = colors, hue = 'Species' )
b3.set_title( "Petal Length for every species", fontsize=14, fontweight='bold' )
b3.set_ylabel( 'Length (cm)' )

plt.subplot(2,2,4)
b4 = sns.boxplot( data = data, x = 'Species', y = 'PetalWidthCm', palette = colors, hue = 'Species' )
b4.set_title( "Petal Width for every species", fontsize=14, fontweight='bold' )
b4.set_ylabel( 'Width (cm)' )

sns.set_style( 'whitegrid' )
plt.suptitle( "Boxplots for each feature categorized by iris species", fontsize=20, fontweight='bold' )
plt.tight_layout()
plt.show()
```
