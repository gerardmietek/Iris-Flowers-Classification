## 4. Label Encoding

This section prepares the dataset for machine learning by performing the following steps:

1. **Feature and Label Separation**:
   - Features are extracted from the first four columns of the dataset.
   - Labels (species) are extracted as a separate column.

2. **Class Distribution**:
   - The distribution of the three species is printed to verify uniformity.
   - Since each species has an equal number of samples, **stratification is not needed** when splitting the data into training and testing sets.

3. **Label Encoding**:
   - The species labels are encoded into numeric values (`0`, `1`, `2`) using `LabelEncoder`, preparing them for use in machine learning models.

This ensures that the labels are in a machine-readable format.


```python
features = data.iloc[:, :4]
labels = data[['Species']]
print( labels.value_counts() )
#no need of stratifing while splitting, because the distribution of every species is uniform
le = LabelEncoder()
labels_encoded = le.fit_transform( labels.values.ravel() )
```

## 5. Splitting the Dataset into Training and Testing Sets

This step divides the dataset into training and testing subsets:

- **Input Features (`X`)**: The dataset's first four columns.
- **Target Labels (`y`)**: The numeric-encoded species labels.

### Key Details:
- The dataset is split using `train_test_split` from `scikit-learn`.
- **Test Size**: 33% of the data is allocated to the testing set, while the remaining 67% is used for training.
- **Stratification**: Not applied because the class distribution is already uniform across species.

After splitting:
- `X_train` and `y_train` represent the training set.
- `X_test` and `y_test` represent the testing set.

The split ensures that the model can be trained and evaluated on separate subsets to assess its performance.


```python
X_train, X_test, y_train, y_test = train_test_split( features, labels_encoded, test_size = 0.33 )
```

This function evaluates a given machine learning model using the following steps:

1. **Model Training**: Fits the model to the training data.
2. **Prediction**: Generates predictions on the test set.
3. **Accuracy Measurement**: Calculates the accuracy score of the model.
4. **Confusion Matrix**: Visualizes the model's performance with a heatmap.
5. **Classification Report**: Prints precision, recall, and F1-score for each class.

```python
def model_tester( model, X_train, X_test, y_train, y_test ):
    #fitting the model
    model.fit( X_train, y_train )

    #predicting values
    y_predicted = model.predict( X_test )

    #measuring the accuracy
    accuracy = accuracy_score( y_test, y_predicted )

    #plotting the confusion matrix
    cm = confusion_matrix( y_test, y_predicted )
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, cmap="Reds", annot=cm, fmt="d", cbar=True, 
                linewidths=0.5)
    print( f"Accuracy score of {model}: ", accuracy_score( y_test, y_predicted) )
    plt.title(f"Confusion Matrix of {model} model")
    plt.xlabel("Predicted Label")
    plt.ylabel("True Label")
    plt.show()

    print(classification_report(y_test, y_predicted))
```

## 6. Training a few simple machine learning algorithms using the previously prepared utility

```python
knn = KNeighborsClassifier()
model_tester( knn, X_train, X_test, y_train, y_test )
```

```python
logistic_regression = LogisticRegression( multi_class = "multinomial", random_state = 42 )
model_tester( logistic_regression, X_train, X_test, y_train, y_test )
```

```python
decision_tree = DecisionTreeClassifier( random_state = 42 )
model_tester( decision_tree, X_train, X_test, y_train, y_test )
```

```python
naive_bayes = GaussianNB()
model_tester( naive_bayes, X_train, X_test, y_train, y_test )
```

This function tunes hyperparameters of a given machine learning model with the following steps:

1. **Grid Search Initialization**: Sets up `GridSearchCV` with the specified parameters.
2. **Model Training**: Fits the model to the training data.
3. **Best Parameters**: Prints the best hyperparameters found during the search.
4. **Best Score**: Displays the best cross-validation score.
5. **Return Best Model**: Returns the model with the optimal hyperparameters.


```python
def model_hyperparameter_tuner(model, X_train, X_test, param_grid, cv=5, random_state = 42):
    # Initialize GridSearchCV
    gs = GridSearchCV(estimator=model, param_grid=param_grid, cv=cv, n_jobs=-1)
    
    # Fit the model to the training data
    gs.fit(X_train, y_train)
    
    # Print the best parameters and the best score
    print("Best hyperparameters:", gs.best_params_)
    print("Best cross-validation score (accuracy):", gs.best_score_)
    
    # Get the best model (estimator)
    best_model = gs.best_estimator_
    
    # Return the best model and test accuracy
    return best_model
```

## 7. Tuning the same machine learning algorithms using the previously prepared utility

```python
dtc_params_grid = { 'criterion': ['gini', 'entropy'],
                    'min_samples_split' : np.arange(2,10),
                    'min_samples_leaf' : np.arange(1,20) }
best_dtc = model_hyperparameter_tuner( DecisionTreeClassifier(random_state = 42), X_train, X_test,  dtc_params_grid, cv=10 )
```

```python
logreg_params_grid = {
    'penalty': ['l2', 'l1'],
    'C': np.logspace(-3, 3, 7),
    'solver': ['liblinear', 'saga'],
    'max_iter': [100, 200, 500, 1000]
}
best_logreg = model_hyperparameter_tuner(LogisticRegression(random_state=42), X_train, X_test, logreg_params_grid, cv=10)
```

```python
gnb_params_grid = {
    'var_smoothing': np.logspace(0, -9, num=100)
}

best_gnb = model_hyperparameter_tuner(GaussianNB(), X_train, X_test, gnb_params_grid, cv=10)
```

```python
knn_params_grid = {
    'n_neighbors': range(1, 31),
    'weights': ['uniform', 'distance'],
    'metric': ['euclidean', 'manhattan', 'minkowski'],
    'leaf_size': [20, 30, 40, 50],
}

best_knn = model_hyperparameter_tuner(KNeighborsClassifier(), X_train, X_test, knn_params_grid, cv=10)
```
